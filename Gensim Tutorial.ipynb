## Preprocess

```python
# Tokenize data: Handling punctuations and lowercasing the text 

from gensim.utils import simple_preprocess 

# preprocess the file to get a list of tokens 
token_list =[] 

for sentence in text.split('.'): 
    
  # the simple_preprocess function returns a list of each sentence 
  token_list.append(simple_preprocess(sentence, deacc = True)) 
    
```

- `deacc` (bool, optional)  Remove accent marks from tokens using deaccent()?

- (The deaccent() function is another utility function, documented at the link, which does exactly what the name and documentation suggest: removes accent marks from letters, so that, for example, `''` becomes just `'e'`.

- simple_preprocess(), per its documentation, to discard any tokens shorter than min_len=2 characters.

## TF-IDF

```python
from gensim import models 
import numpy as np 

# create TF-IDF model 
tfIdf = models.TfidfModel(BoW_corpus, smartirs ='ntc') 
```

```python
print(tfIdf)
```

```python
my_dictionary[231]
```

```python
# TF-IDF Word Weight 
weight_tfidf =[] 
for doc in tfIdf[BoW_corpus]: 
    for id, tf_idf in doc: 
        weight_tfidf.append([my_dictionary[id], np.around(tf_idf, decimals = 3)]) 
print(weight_tfidf) 
```

## Doc2Vec

```python
documents=text.split('.')
```

```python
documents
```

```python
from collections import namedtuple

# Transform data (you can add more data preprocessing steps) 
docs = []
analyzedDocument = namedtuple('AnalyzedDocument', 'words tags')
for i, text in enumerate(documents):
    words = text.lower().split()
    tags = [i]
    docs.append(analyzedDocument(words, tags))

print(docs[:2])
```

```python
from gensim.models import doc2vec

model = doc2vec.Doc2Vec(docs,
                        vector_size=100,
                        window=2, 
                        min_count=1, 
                        workers=4,
                        dm=0) # PV-DBOW
```

- dm ({1,0}, optional)  Defines the training algorithm. If dm=1, distributed memory (PV-DM) is used. Otherwise, distributed bag of words (PV-DBOW) is employed.

- vector_size (int, optional)  Dimensionality of the feature vectors.

- window (int, optional)  The maximum distance between the current and predicted word within a sentence.

- alpha (float, optional)  The initial learning rate.

```python
model.infer_vector(['the', 'tragedy', 'of', 'hamlet,', 'prince', 'of', 'denmark', 'by', 'william', 'shakespeare', 'dramatis', 'personae', 'claudius,', 'king', 'of', 'denmark'])
```

## FastText

```python
# Import FastText 
from gensim.models import FastText

# Create FastText Model object
model = FastText(size=50, window=3, min_count=1)  # instantiate

# Build Vocab
model.build_vocab(sentences=token_list)

# Train FastText model
model.train(sentences=token_list, total_examples=len(token_list), epochs=10)  # train
```

```python
Feature engineering 
Transofrm text into vectors. 

context-free models 
contextual feature models.

Frequency Based Features: BoW and TF-IDF


BoW tokenized sentences/paragraphs into words and counts the occurrence of words.

TFIDF
Term Frequency considers all the terms or words with equal importance but some words such as helping verbs and articles are more frequent than other words and such words are less important to classify the documents.

Another measure Inverse Document Frequency (IDF) measures the importance of words. 
Which measures how important a term is? IDF assigns more weight to rare terms and less weight to frequent terms 


Prediction Based Features
neural network language model (NNLM) is trained on it.
Continuous Bag of Words (CBOW) predicts the current word based on four future and four history words using the log-linear classifier. 
Skip-gram predicts the context of the given word. Skip-gram takes the current word as input and predicts the before and after the current word using a log-linear classifier with a continuous projection layer.

Each word vector is concatenated as a single vector to predict the next word. This model is known as the Distributed Memory Model of Paragraph Vectors (PV-DM) because paragraph tokens remember the missing context in the paragraph. 

Another feature engineering method, the Distributed Bag of Words version of Paragraph Vector (PV-DBOW) ignores the context of input words and predicts missing words from a random sample. 
```
